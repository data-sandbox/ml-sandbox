## Clustering

Sandbox to experiment with clustering methods.

Resources:
- https://machinelearningmastery.com/clustering-algorithms-with-python/

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_classification
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set()
```

```python
# Generate sythetic data
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2,
                           n_redundant=0, n_classes=3, n_clusters_per_class=1, random_state=2)

# Inspect
plt.figure(figsize=(4, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
plt.show()
```

### K-Means

```python
# Define model
model = KMeans(n_clusters=3, n_init='auto')

# Fit, predict
model.fit(X)
y_pred = model.predict(X)

# Inspect
plt.figure(figsize=(4, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_pred)
plt.show()
```

A reasonable grouping is found. However, one downside of K-means here is that it does not perform great when there are significant unequal variances between groups like in this case.

### Gaussian Mixture Model

```python
model = GaussianMixture(n_components=3)

# Fit, predict
model.fit(X)
y_pred = model.predict(X)

# Inspect
plt.figure(figsize=(4, 4))
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y_pred)
plt.show()
```

The Gaussian Mixture Model appears to cluster the data nearly perfectly! This can be explained by the fact that the synthetic data was actually generated by a mixture of Gaussians.
